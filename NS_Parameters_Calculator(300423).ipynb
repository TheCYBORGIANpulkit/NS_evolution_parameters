```python
#calculation of dipole radius and dipole moment
#calculation of radius in SI units, calibrated using R = 11.2 Km for M = 1.4 Msun
radius = 1.23247e+14*(m2**(-0.33))
#taking B = 1.5e+4 Tesla = 1.5e+8 Gauss
dipole_moment = (1.5*(10**4))*(radius**3)
#print(radius)
#print(dipole_moment)
#calculation of r_m
G = 6.67*(10**(-11))
c = 3.0*(10**8)
pi = math.pi
r_m = [0]*len(model)
inner_rad = 6*G*m2/(c**2)
r_mlimit = [0]*len(model)
for i in range(len(model)):
    r_m[i] = 1*((dipole_moment[i]**4)/(2*G*m2[i]*(m2dot_b[i]**2)))**(1/7)
    if(i == 3000): print("....",model[i],".....", r_m[i], "........")
    r_mlimit[i] = max(inner_rad[i], radius[i])
    if(r_m[i] <= r_mlimit[i]):
        r_m[i] = r_mlimit[i]
    print(model[i],"...",r_m[i],"...",m2dot[i],"....",r_mlimit[i],"...", dipole_moment[i], "...", radius[i], "...." , inner_rad[i])  
print(model[3000], "...." , r_m[3000], "....", r_mlimit[3000], "...." , dipole_moment[3000])
#print(r_m[0])
print(len(model))
```

```python
#define initial spin frequency and initilaie the arrays 
nu = [1.0]*len(model)
r_lc = [0.0]*len(model)
r_co = [0.0]*len(model)
N_acc = [0.0]*len(model)
N_field = [0.0]*len(model)
N_EM = [0.0]*len(model)
N_GW = [0.0]*len(model)
N = [0.0]*len(model)
J = [1.0]*len(model)
delta_J = [0.0]*len(model)
nu_eq = [1.0]*len(model)
#print(J)
#now, calculation of r_{lc} and r_{co} can only be done using the for loop
for i in range(len(model)):
    #calculation of timestep in seconds
    if (i == 0):
        dt = age[i]*3.154e+7
        r_lc[i] = c/(2*pi*nu[i])
        r_co[i] = ((G*m2[i])/(4*(pi**2)*(nu[i]**2)))**(1/3)
    else:
        dt = (age[i] - age[i - 1])*3.154e+7
        r_lc[i] = c/(2*pi*nu[i-1])
        r_co[i] = ((G*m2[i])/(4*(pi**2)*(nu[i-1]**2)))**(1/3)
    #calculation of torque N = N_acc + N_field in
    #accretion phase
    if(r_m[i] < r_co[i]):
        N_acc[i] = m2dot_b[i]*math.sqrt(G*m2[i]*r_m[i])
        #eta = 1. 
        N_field[i] = ((dipole_moment[i]**2)/(9*(r_m[i]**3)))*(2*((r_m[i]/r_co[i])**3) - 6*((r_m[i]/r_co[i])**(3/2)) + 3)
        if(r_m[i] == r_mlimit[i]):
            N[i] = N_acc[i] 
        else:
            N[i] = N_acc[i] + N_field[i] 
    #propeller phase
    else:
        N_acc[i] = - m2dot_b[i]*math.sqrt(G*m2[i]*r_m[i])
        N_field[i] = - ((dipole_moment[i]**2)/(9*(r_m[i]**3)))*(3 - 2*((r_co[i]/r_m[i])**(3/2)))
        #N_EM[i] = 
        #N_GW[i] = 
        if(r_m[i] == r_mlimit[i]):
            N[i] = N_acc[i] 
        else:
            N[i] = N_acc[i] + N_field[i]
    #print(N)
    #updating J and nu. 
    #calculate J
    J[i] = 0.4*m2[i]*(radius[i]**2)*2*pi*nu[i-1]
    delta_J[i] = N[i]*dt
    #print(delta_J[i])
    
    J[i] = J[i] + delta_J[i]
    #print("...", r_m[i], ",", r_co[i] , ",", N, "," ,  J[i], "...")
    #print(J[i])
    nu[i] = J[i]/(0.4*m2[i]*(radius[i]**2)*2*pi)
    #nu_eq[i] = (1/((2**0.78571428571)*pi*(1**1.5)))*(((G**5)*(m2[i]**5)*(m2dot_b[i]**3))/(dipole_moment[i]**6))**(1/7)
    nu_eq[i] = 0.18464032643*(((G**5)*(m2[i]**5)*(m2dot_b[i]**3))/(dipole_moment[i]**6))**(1/7)
    #nu_eq[i] = (1/2*pi)*math.sqrt((G*m2[i])/(r_co[i]**3))
    print(model[i]," ", age[i], " ",r_m[i]," ",r_co[i], " ", nu[i], " ", nu_eq[i])

print("NO Errors until now")
print(len(model))
print(len(nu))
```
